## Outliers

```python
import pandas as pd
ser=pd.Series(payment)
ser
```

```python
#1st Quantile(median of first half)
q1=ser.quantile(0.25)
print('Q1:',q1)

#3rd Quantile(median of second half)
q3=ser.quantile(0.75)
print('Q3:',q3)

#Find the IQR (inter quantile range) to identify outliers
iqr=q3-q1
print('IQR:',iqr)
```

## Outliers Detection Formula

##### higher side==> Q3 + (1.5 * IQR), lower side ==> Q1 - (1.5 * IQR)

```python
upside=q3 + (1.5 *iqr)
print(upside)
```

```python
#Anything above 68.5 is outlier
```

```python
lowerside=q1 - (1.5 *iqr)
print(lowerside)
```

```python
#Anything below -17.5 is outlier
```

## Categorical Data Plotting

```python
#--> Catplot() --> Boxplot() --> Striplot() --> Swarmplot()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#catplot = reate several types of plots like box plots, strip plots, swarm plots, and more, depending on the kind parameter.
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#swarmplot = similar to the stripplot, but the points are arranged to avoid overlap by adjusting their positions. It displays individual data points like stripplot but with better spacing for clarity.
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#another way to print swarm
#hue adds color based on size variable in table tips
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
plt.show()
```

```python
#stripplot() = Displays individual data points for a categorical variable, using a scatter plot along a categorical axis.
```

```python
sns.stripplot(x='size',y='total_bill',data=tips)
plt.show()
```

```python
#different x label
sns.stripplot(x='smoker',y='total_bill',data=tips)
plt.show()
```

## Linear Regression and Relationship

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#regplot() = Regression plot
#A function for creating a scatter plot with a linear regression line fitted to the data, optionally with confidence intervals.
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#lmplot() = Linear Model Plot
#lmplot() allows you to fit regression models to data and visualize the relationship between variables with additional functionality like faceting.
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```
