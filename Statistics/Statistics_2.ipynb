## Probablity Problem
Problem 1 - There are 1000 Students whose marks are captured based on their ability in the test if the same test given to a new student . what is the probablity of getting 50 marks.

```python
from  numpy.random import randint as ri
import pandas as pd
```

```python
#generate 1000 random int between 20-101
marks=ri(20,101,1000)
```

```python
marks=pd.DataFrame(marks)
marks.head()
```

```python
#Lets calculate mean and std

marks_mean=marks.mean()
print('Mean of marks: ',marks_mean)
marks_std=marks.std()
print('Standard Deviation of marks: ',marks_std)
```

```python
#what is the probablity of student scoring upto 50% marks
```

```python
x=50
```

```python
#probability always between 0-1 and starts from 0
```

```python
import scipy.stats as st
#when x(data) is given then, use CDF function to find the probability of scoring marks upto 50(x)
```

```python
#st: alias for the scipy.stats module (imported as import scipy.stats as st).
#norm is the normal distribution class.
#cdf stands for Cumulative Distribution Function, which calculates the probability that a random variable is less than or equal to a given value x.
#x: This is the value for which you want to find the cumulative probability.
#loc=marks_mean: loc specifies the mean of the normal distribution. In your case, it is the mean of the marks (marks_mean).
#scale=marks_std: scale specifies the standard deviation of the normal distribution. Here, it is the standard deviation of the marks (marks_std).


prob=st.norm.cdf(x,loc=marks_mean,scale=marks_std)
print(prob)
```

```python
#If You Want to Find The Probablity Of Scoring Marks Greater then 50, you can minus prob(1-prob)
```

#### Alternate Hypothesis -Ha-> Alternative hypothesis is the hypothesis that we are trying to prove and which is accepted if we have sufficient evidence to reject the null hypothes, It isDecision leads to opposite of H0.

##### Problem 1 -
A car manufacturing company claims that the mileage of their cars is 25KMPL . The population standard deviation is known from the past experiance to be 1.5KMPL. A car rental company wants to test the null hypothesis that the average mileage is at least 25KMPL. Car rental representative drove 30 cars (Samples) of the same company and record each car's mileage. The average mileage of the sample turned out to be 24.2 KMPL.

Test the null hypothesis ? (H0) Null Hypothesis >=25 (H1) Alternate Hypothesis <=25

```python
import numpy as np

#given values
mean=25
sample30_mean=24.2
std=1.5
n=30
```

```python
#convert pop std ---> sample std 
#sample std = pop std/sqrt(no of sample)
z=(sample30_mean-mean)/(std/np.sqrt(30))
print(z)
```

```python
p_value=st.norm.cdf(sample30_mean,loc=mean,scale=1.5/np.sqrt(30))
print(p_value) #also can check from z distribution table
```

## Import points
-> Standard alpha level =0.05

-> if p-Value is < 0.05 then we reject null hypothesis (Ho)

-> if p values is > 0.05 then we failed to reject the Ho. We reject alternate hypothesis.

-> p value is also called alpha value

Conclusion- As the p value is less than alpha (0.05), we reject the null hypothesis (mileage>=25KMPL) not given 25 mileage so rental company will not buy the cars.

## Type 1 Error -> If Ho is True and You reject Ho
In our exapmple, if average car mileage is really 25 or >25 KMPL and rental agency rejects it. Its type 1 Error.

## Type 2 Error -> If Ho is False and you failed to reject Ho
If average car mileage is <25 but rental agency is buying cars (failed to reject Ho). Its type 2 Error.
