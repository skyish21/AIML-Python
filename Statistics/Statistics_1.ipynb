```python
#2-Measure of Dispersion(Spread)(Variance and Standard Deviation) - how spread out or varied the data is.

#Variance-How much a value differs from average or mean
```

```python
#Variance
import seaborn as sns

lst=[1,2,3,4,5,6,7,8,9,1,2,3,4]
sns.scatterplot(lst)
```

```python
#Standard Deviation
mean=np.mean(lst)
print('Mean: ',mean)
sns.distplot(lst)
```

### 

```python
#For Calculation of Variance there is a formula ^ .
#Measure of dispersion
#x=data,Mu is mean, N= occurance
```

```python
import numpy as np

lst=[21,23,34]
avg=np.mean(lst)
print(avg)
```

```python
#first sq the difference then add to other values then divide by total number of elements then sum it all
result=np.sum((np.square(21-avg)+np.square(23-avg)+np.square(34-avg))/3)
print(result)
```

```python
std_dev=np.sqrt(result)
print(std_dev)
```

```python
#directly find variance and standard deviation

print(np.var(lst))
```

```python
print(np.std(lst))
```
