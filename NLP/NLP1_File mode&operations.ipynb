### Import Text

In NLP, dataset will be in .txt, .csv format. You need to import and clean the irrelevant data from your dataset.

Operations:
- create
- open
- append
- write

### Create text file

You can create text file in python using open(). 

```python
# 2 arguments: filename, permission/operation to be performed on file
# w: write, r: read, a: append
# It will create a new file if it doesn't already exist.

file = open("datasetfile.txt" , "+w")
```

```python
# Loop runs 5 iterations
# Using wrtie function we enter data into the file
#%d: Placeholder for an integer.
#\r\n: A carriage return and newline, used for Windows-style line breaks (only \n would work for Unix-like systems).
#% (i+1): Replaces %d with i + 1, so the line numbers start from 1 instead of 0.

for i in range(5):
    file.write("Line number: %d\r\n" % (i+1))
```

```python
# This will close the file
file.close()
```

### Append the file

Add data to existing file or new file.

```python
# a+: Append and read mode.
file = open("datasetfile.txt", "a+")
```

```python
for i in range(3):
   file.write("Appending Line number: %d\r\n" % (i+1))
```

```python
file.close()
```

### Read the files

```python
file = open("datasetfile.txt", "r")
```

```python
if file.mode == "r":
    content = file.read() # Save contents of the file in content variable.
```
